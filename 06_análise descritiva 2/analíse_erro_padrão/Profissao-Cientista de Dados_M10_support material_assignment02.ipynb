````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
renda = pd.read_csv('previsao_de_renda.csv')
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=renda, dodge=True, errorbar=('ci', 90))
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=renda, dodge=True, errorbar=('ci', 90))
````

- Ambas as variáveis impactam extremamente a renda. E, conforme visto nos gráficos acima, a discrepância indica que isso não só afetaria a amostra como também a população como um todo.
