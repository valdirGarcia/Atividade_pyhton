#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter

%matplotlib inline

sinasc = pd.read_csv('SINASC_RO_2019.csv')

sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(18, 12))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[0, 2], x='SEXO', data=sinasc)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=sinasc)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=sinasc)

axes[1, 0].tick_params(axis="x", rotation=30)  
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc)
plt.xlabel('Tipo de Consulta')
plt.ylabel('APGAR5 Médio')
plt.title('APGAR5 Médio por Tipo de Consulta')

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(18, 12))
fig, axes1 = plt.subplots(2, 2)

sns.countplot(ax = axes1[0, 0], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes1[0, 1], x='CONSULTAS', data=sinasc)
sns.countplot(ax = axes1[1, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes1[1, 1], x='GESTACAO', data=sinasc)

axes1[0, 0].yaxis.set_major_formatter(PercentFormatter(xmax=len(sinasc)))
axes1[0, 1].yaxis.set_major_formatter(PercentFormatter(xmax=len(sinasc)))
axes1[1, 0].yaxis.set_major_formatter(PercentFormatter(xmax=len(sinasc)))
axes1[1, 1].yaxis.set_major_formatter(PercentFormatter(xmax=len(sinasc)))

axes1[1, 1].tick_params(axis="x", rotation=30)  
```
